```python
#TASK2
#The cost column in the dataframe “Project” has some missing values. Your task is to compute these missing values.
#Replace the missing values by running average. You should use the “For” loop for this task.



import pandas as pd

# Read CSV file into dataframe
project_df = pd.read_csv('project_data.csv')

# Convert 'Cost' column to numeric
project_df['Cost'] = pd.to_numeric(project_df['Cost'], errors='coerce')

# Compute running average and replace missing values
running_sum = 0
count = 0
for index, row in project_df.iterrows():
    if pd.isnull(row['Cost']):
        project_df.at[index, 'Cost'] = running_sum / count if count > 0 else 0
    else:
        running_sum += row['Cost']
        count += 1

# Save dataframe with filled missing values
project_df.to_csv('project_data_filled.csv', index=False)

# Display updated dataframe
print(project_df)
```

```python
#TASK3 
#Split the name column in the Employee dataframe into two new 
# columns “First Name”, and “Last Name” and remove the older “name” column.


import pandas as pd

# Read CSV file into dataframe
employee_df = pd.read_csv('employee_data.csv')

# Split 'Name' column into 'First Name' and 'Last Name'
employee_df[['First Name', 'Last Name']] = employee_df['Name'].str.split(' ', n=1, expand=True)

# Drop the old 'Name' column
employee_df.drop(columns=['Name'], inplace=True)

# Save updated dataframe to a new CSV file
employee_df.to_csv('employee_data_updated.csv', index=False)

# Display updated dataframe
print(employee_df)
```

```python
#TASK4
#Join all three dataframes in one single dataframe. Name it “Final”


import pandas as pd

# Read CSV files into dataframes
project_df = pd.read_csv('project_data.csv')
employee_df = pd.read_csv('employee_data.csv')
seniority_df = pd.read_csv('seniority_data.csv')

# Merge dataframes
merged_df = pd.merge(project_df, employee_df, on='ID')
merged_df = pd.merge(merged_df, seniority_df, on='ID')

# Rename the columns to make them more informative
merged_df.rename(columns={'ID': 'Employee ID'}, inplace=True)

# Save merged dataframe to a new CSV file
merged_df.to_csv('final_data.csv', index=False)

# Display the merged dataframe
print(merged_df)
```

```python
#TASK5  
#Add a new bonus column in the Final dataframe.
#Give a 5% bonus concerning project cost only to employees who have finished the projects.



import pandas as pd

# Read CSV files into dataframes
project_df = pd.read_csv('project_data.csv')
employee_df = pd.read_csv('employee_data.csv')
seniority_df = pd.read_csv('seniority_data.csv')

# Merge dataframes
merged_df = pd.merge(project_df, employee_df, on='ID')
merged_df = pd.merge(merged_df, seniority_df, on='ID')

# Convert 'Cost' column to numeric
merged_df['Cost'] = pd.to_numeric(merged_df['Cost'], errors='coerce')

# Calculate bonus based on project status
merged_df['Bonus'] = 0  # Initialize bonus column with 0

# Apply bonus only to employees who have finished their projects
finished_projects = merged_df[merged_df['Status'] == 'Finished'].copy()  # Make a copy to avoid the SettingWithCopyWarning
finished_projects.loc[:, 'Bonus'] = finished_projects['Cost'] * 0.05

# Update bonus column in the merged dataframe
merged_df.update(finished_projects)

# Save merged dataframe with bonus to a new CSV file
merged_df.to_csv('final_data_with_bonus.csv', index=False)

# Display the updated dataframe
print(merged_df)

```

```python
#Task 6
#Demote the designation level by 1, whose projects have status “fail”. 
#Delete the employees record whose designation level is above 4.

import pandas as pd

# Read CSV files into dataframes
project_df = pd.read_csv('project_data.csv')
employee_df = pd.read_csv('employee_data.csv')
seniority_df = pd.read_csv('seniority_data.csv')

# Merge dataframes
merged_df = pd.merge(project_df, employee_df, on='ID')
merged_df = pd.merge(merged_df, seniority_df, on='ID')

# Convert 'Designation Level' column to numeric
merged_df['Designation Level'] = pd.to_numeric(merged_df['Designation Level'], errors='coerce')

# Demote designation level by 1 for employees whose projects have a status of "Failed"
merged_df.loc[merged_df['Status'] == 'Failed', 'Designation Level'] -= 1

# Delete records of employees whose designation level is above 4
merged_df = merged_df[merged_df['Designation Level'] <= 4]

# Save the updated dataframe to a new CSV file
merged_df.to_csv('final_data_updated.csv', index=False)

# Display the updated dataframe
print(merged_df)




```

```python
#TASK 7
#Add “Mr.” and “Mrs.” to the first name column and drop the gender column.


import pandas as pd

# Read CSV files into dataframes
project_df = pd.read_csv('project_data.csv')
employee_df = pd.read_csv('employee_data_updated.csv')  # This file has split first name and last name
seniority_df = pd.read_csv('seniority_data.csv')

# Merge dataframes
merged_df = pd.merge(project_df, employee_df, on='ID')

# Add "Mr." and "Mrs." to the first name column based on gender
merged_df['First Name'] = merged_df.apply(lambda row: f"Mr. {row['First Name']}" if row['Gender'] == 'M' else f"Mrs. {row['First Name']}", axis=1)

# Drop the gender column
merged_df.drop(columns=['Gender'], inplace=True)

# Save the updated dataframe to a new CSV file
merged_df.to_csv('final_data_with_titles.csv', index=False)

# Display the updated dataframe
print(merged_df)

```

```python
#TASK8
#Promote designation level by 1 for the employees whose age is more than 29 years using IF condition.

import pandas as pd

# Read CSV files into dataframes
project_df = pd.read_csv('project_data.csv')
employee_df = pd.read_csv('employee_data_updated.csv')
seniority_df = pd.read_csv('seniority_data.csv')

# Merge dataframes
merged_df = pd.merge(project_df, employee_df, on='ID')
merged_df = pd.merge(merged_df, seniority_df, on='ID')

# Promote designation level by 1 for employees whose age is more than 29 years
merged_df.loc[merged_df['Age'] > 29, 'Designation Level'] += 1

# Save the updated dataframe to a new CSV file
merged_df.to_csv('final_data_promoted.csv', index=False)

# Display the updated dataframe
print(merged_df)
```

```python
#TASK9
#Add the cost of all projects for each Employee and save it in new dataframe
#“TotalProjCost” with three columns ID, First Name, and Total cost


import pandas as pd

# Read CSV files into dataframes
project_df = pd.read_csv('project_data.csv')
employee_df = pd.read_csv('employee_data_updated.csv')
seniority_df = pd.read_csv('seniority_data.csv')

# Merge dataframes
merged_df = pd.merge(project_df, employee_df, on='ID')
merged_df = pd.merge(merged_df, seniority_df, on='ID')

# Convert 'Cost' column to numeric
merged_df['Cost'] = pd.to_numeric(merged_df['Cost'], errors='coerce')

# Group by 'ID' and 'First Name', and sum up the costs of their projects
total_proj_cost_df = merged_df.groupby(['ID', 'First Name']).agg({'Cost': 'sum'}).reset_index()

# Rename the column for clarity
total_proj_cost_df.rename(columns={'Cost': 'Total Cost'}, inplace=True)

# Save the new dataframe to a new CSV file
total_proj_cost_df.to_csv('TotalProjCost.csv', index=False)

# Display the new dataframe
print(total_proj_cost_df)

```
